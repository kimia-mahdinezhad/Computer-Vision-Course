```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from numba import jit
from sklearn.metrics import mean_squared_error
```

```python
img = cv2.imread("Barbara.bmp", 0)
```

```python
@jit(nopython=True)
def change_data_bit(image, input_bit):
  shift_bit = 8 - input_bit

  return image>>shift_bit
```

```python
fig, plot = plt.subplots(1, 1, figsize = (10, 5))

plot.imshow(img, cmap='gray')
plot.set_title("Image: Original")
```

```python
input_bits = [5, 4, 3, 2, 1]

fig, plot = plt.subplots(1, len(input_bits), figsize = (20, 20))

for input_bit_index in range (len(input_bits)):
  shifted_image = change_data_bit(img, input_bits[input_bit_index])

  plot[input_bit_index].imshow(shifted_image, cmap='gray')
  plot[input_bit_index].set_title("Image: Shifted (Input Bit = " + str(input_bits[input_bit_index]) + ")")
```
